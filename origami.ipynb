【問題1】べき乗の算術演算子を使用した実装

```python
# 紙を1回折った時の厚さを計算するコード

THICKNESS = 0.00008
folded_thickness = THICKNESS * 2 ** 43
print("厚さ： {}メートル".format(folded_thickness))
```

【問題2】単位の変換

```python
# メートルをキロメートルに変換して小数点以下2桁で表示する
print("厚さ： {:.2f}キロメートル".format(folded_thickness / 1000))
```

【問題3】for文を使用して作成

```python
THICKNESS = 0.00008
for i in range(int(43)):
    THICKNESS = THICKNESS * 2
print(THICKNESS)
```

【問題4】計算時間の比較

```python
#時間計測①
import time
start = time.time()
#####
THICKNESS = 0.00008
folded_thickness = THICKNESS * 2 ** 43
print("厚さ： {}メートル".format(folded_thickness))
#####
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))


```

```python
#時間計測②
import time
start = time.time()
#####
THICKNESS = 0.00008
for i in range(int(43)):
    THICKNESS = THICKNESS * 2
print(THICKNESS)
#####
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

```python
#繰り返し時間計測①
import time
for i in range(int(5)):
    start = time.time()
    THICKNESS = 0.00008
    folded_thickness = THICKNESS * 2 ** 43
    print("厚さ： {}メートル".format(folded_thickness))
    elapsed_time = time.time() - start
    print("time : {}[s]".format(elapsed_time))
```

【問題5】リストへの保存

```python
THICKNESS_list = []
THICKNESS = float(0.00008)
THICKNESS_list.append(THICKNESS)
for i in range(int(43)):
    THICKNESS = THICKNESS * 2
    THICKNESS_list.append(THICKNESS)
print(THICKNESS_list)
print(len(THICKNESS_list))
```

【問題6】折れ線グラフの表示

```python
import matplotlib.pyplot as plt
%matplotlib inline
# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(THICKNESS_list)  # 「リスト名」のところにリストの変数名を入れる
plt.show()

#説明　X軸が紙を折った階数、Y軸が紙の厚さ。紙を折る階数に比例して紙の厚さが指数関数的に増えている
```

【問題7】グラフのカスタマイズ

```python
##①軸のラベルサイズが大きいグラフ---------------------------------------------------
import matplotlib.pyplot as plt
%matplotlib inline
# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("①thickness of folded paper LABEL SIZE20")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.tick_params(labelsize=20) # 軸の値に関する設定を行う
plt.plot(THICKNESS_list)  # 「リスト名」のところにリストの変数名を入れる
plt.show()

##②線の色が赤いグラフ--------------------------------------------------------------
import matplotlib.pyplot as plt
%matplotlib inline
# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("②thickness of folded paper RED LINE")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(THICKNESS_list, color='red')  # 「リスト名」のところにリストの変数名を入れる
plt.show()

##③グリッド付のグラフ--------------------------------------------------------------
import matplotlib.pyplot as plt
%matplotlib inline
# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("③thickness of folded paper with GRID")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.grid(True)
plt.plot(THICKNESS_list)  # 「リスト名」のところにリストの変数名を入れる
plt.show()

```
