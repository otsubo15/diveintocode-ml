###  【問題6】ラベル付け

```python
#----インポート--------------------------------------------
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

#----データ①----------------------------------------------

#平均値
mean_1 = np.array([-3, 0])
#共分散行列
cov_1 = np.array([[1, 0.8], [0.8, 1.0]])

#乱数①の作成（乱数のseedは0）
np.random.seed(0)
random_data_1 = np.random.multivariate_normal(mean_1, cov_1, size = 500)
#print(random_data_1)

#ランダムデータ①の0カラムをXデータとしてスライス
X_1 = (random_data_1[:, 0])
#print(X)

#ランダムデータ①の1カラムをYデータとしてスライス
Y_1 = (random_data_1[:, 1])
#print(Y)

#----データ②------------------------------------------------

#引数
mean_2 = np.array([0, -3]) #平均値
cov_2 = np.array([[1.0, 0.8], [0.8, 1.0]]) #共分散行列

#乱数②の作成（乱数のseedは1）
np.random.seed(1)
random_data_2 = np.random.multivariate_normal(mean_2, cov_2, size = 500)
#print(random_data_2)

#ランダムデータ②の0カラムをXデータとしてスライス
X_2 = (random_data_2[:, 0])
#print(X)

#ランダムデータ②の1カラムをYデータとしてスライス
Y_2 = (random_data_2[:, 1])
#print(Y)

#----データ③------------------------------------------------

#データ①とデータ②を結合
random_data_3 = np.concatenate((random_data_1, random_data_2), axis=0)
#print(random_data_3)

# データ型の確認
#print(random_data_3.shape)


#----ラベル付け------------------------------------------------

#500個連続した「0」と「1」のndarrayを作成
zero_ndarray = np.zeros(500) 
ones_ndarray = np.ones(500)
#print(zero_ndarray)
#print(ones_ndarray)

#500個連続した「0」と「1」のndarrayを結合（但し１次元のndarray）
rabel_data_1 = np.concatenate((zero_ndarray, ones_ndarray), axis=0)
#print(rabel_data_1)

#データ型の確認
#print(rabel_data_1.shape)

#２次元のデータ③に１次元のラベルデータ①を挿入（追加）する
print(np.insert(random_data_3, 2, rabel_data_1, axis=1))
```
