【問題】栗まんじゅうが太陽系を覆う日

```python
import matplotlib.pyplot as plt
%matplotlib inline

"""---------前提条件（Prerequisites）---------"""
kuri_vol = 10 ** -4 #栗まんじゅうの体積　単位は(ｍ3)
space_vol = (27/2) * 10 ** 78 #宇宙の体積　単位は(ｍ3)

###------------------------------
Target_object_list = []
Arrival_time_list = []

def BaiBain (Doubled_object,Target_object):
    while Target_object >= Doubled_object:
        Doubled_object = Doubled_object * 2
        Target_object_list.append(Doubled_object)
        Arrival_time = len(Target_object_list) * 5 #単位は(min)
        Arrival_time_list.append(Arrival_time)
    #print(Target_object_list)
    #print(Arrival_time_list)
    #print(space_vol)
    
    print("倍になる物体が目標の物体を覆うのは『{0}』分後です。".format(Arrival_time))
    #グラフを書く------------------------------
    plt.rcParams["figure.figsize"] = (30, 6)
    plt.title("BaiBain")
    plt.xlabel("Time[min]")
    plt.ylabel("Volume [m3]")
    plt.plot(Arrival_time_list, Target_object_list, marker="o", linestyle='--')
    plt.grid(True)
    plt.show()
    return 

BaiBain(kuri_vol,space_vol)
```
