## 【問題1】2×2マスのチェス盤の小麦の数
2×2マスのチェス盤をndarrayによって表現し、各マスの小麦の数を計算してください。左上のマスに小麦を1つ置いた後、その右が2つ、左下が4つ、右下が8つという風に進みます。
サンプルコードで作成したndarrayのshapeの変形を行ってください。

```python
import numpy as np
n_squares = 4 #2×2マスだと総マス数は4マス
small_board_list = [1] #最初の１粒を置く

#---４マス分小麦を置くfor文---
for _ in range(n_squares - 1):
    small_board_list.append(2 * small_board_list[-1])

small_board_ndarray = np.array(small_board_list)
print("4マスの板に小麦を並べる（リスト）：{}".format(small_board_list))

#---サンプルコードで作成したndarrayのshapeの変形---
small_board_ndarray_reshape = np.reshape(small_board_ndarray, (2, 2))
print("4マスの板に小麦を並べる（リスト）：{}".format(small_board_ndarray_reshape))
```

## 【問題2】n×mマスへの拡張
n×mマスのチェス盤の小麦の数を記したndarrayを返す関数を作成してください。そして、この関数を使い8×8マスのチェス盤の小麦の数を求めてください。

## 【問題3】小麦の数の合計
8×8マスのチェス盤の小麦の数の合計を求めてください。また、各列の平均を求め、それを棒グラフで可視化してください。
棒グラフはサンプルコードを用意しています。

```python
#----インポート----
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

#----引数の定義----
#n_Horizontal:横のマス数（列）
#n_Vertical:縦のマス数（行）

board_list = [1] #最初の１粒を置く

#---n×mマスのチェス盤の小麦の数の合計を算出する関数---
def choice_squaresnumber_komugi_chess(n_Horizontal, n_Vertical):
    
    #マス数を計算する（最初の１粒は置いてあるので１引いてある）
    n_squares = int(n_Horizontal) * int(n_Vertical) -1
    
    for i in range(n_squares): 
        
        #最後の粒を２倍してリストに追加する
        board_list.append(2 * board_list[-1]) 
        
    #リストをndarrayに変換
    board_ndarray = np.array(board_list)
    #print("検証中")
    #print(board_ndarray)
    #print("検証中")
    #ndarrayを『n_Horizontal』 x 『n_Horizontal』の配列にする
    board_ndarray_reshape = np.reshape(board_ndarray, (n_Horizontal, n_Vertical))   
    #print("検証中")
    #print(board_ndarray_reshape)
    #print("検証中")
    
    #各行の平均値を入れる為のndarrayを作る
    colums_ave_list = []
    
    #各列の平均値をfor文でリストにしてndarrayにする
    for i in range(int(n_Horizontal)):
        
        #np.mean():で平均を求める[:, i]でi列を指定する。
        colums_ave = np.mean(board_ndarray_reshape[:, i])
        #print(colums_ave)
                
        colums_ave_list.append(colums_ave) ###ここはリストにしたがndarrayで作るにはどうすればよかったか
        colums_ave_ndarray = np.array(colums_ave_list) #リストをndarrayに変換
        
    #各列の平均値の棒グラフ
    plt.xlabel("Index number")
    plt.ylabel("Average value of wheat")
    plt.title("Average value of wheat in each row")
    plt.bar(np.arange(1, n_Horizontal+1), colums_ave_ndarray)
    plt.show()
    
    print("{0}×{1}マスのチェス盤の小麦の数の合計の数は{2}粒です。".format(n_Horizontal, n_Vertical, np.sum(board_ndarray)))
    return

###----関数に縦8マス横8マスを代入----
choice_squaresnumber_komugi_chess(8, 8)
```

## 【問題5】後半は前半の何倍か
チェス盤の後半にある小麦の数は前半にある小麦の数の何倍かを求めてください。前半とは配列の0から3行目、後半は4から7行目です。

```python
#----インポート----
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

#----引数の定義----
#n_Horizontal:横のマス数
#n_Vertical:縦のマス数

board_list = [1] #最初の１粒を置く

#---n×mマスのチェス盤の小麦の数の合計を算出する関数---
def choice_squaresnumber_komugi_chess(n_Horizontal, n_Vertical):
    
    #マス数を計算する（最初の１粒は置いてあるので１引いてある）
    n_squares = int(n_Horizontal) * int(n_Vertical) -1
    
    #マス数だけ小麦を２倍にして繰り返し置いていく
    for i in range(n_squares): 
        
        #最後の粒を２倍してリストに追加する
        board_list.append(2 * board_list[-1]) 
        
    #リストをndarrayに変換
    board_ndarray = np.array(board_list)
    
    #ndarrayを『n_Horizontal』 x 『n_Horizontal』の配列にする
    board_ndarray_reshape = np.reshape(board_ndarray, (n_Horizontal, n_Vertical))   
    
    #0-3の合計値（index_sum_0_3）を求める
    index_sum_0_3 = 0
    for i in range(0,4):
        index_sum_0_3 = index_sum_0_3 + np.sum(board_ndarray_reshape[i])
    #print(index_sum_0_3)
    
    #4-7の合計値（index_sum_4-7）を求める
    index_sum_4_7 = 0
    for i in range(5,8):
        index_sum_4_7 = index_sum_4_7 + np.sum(board_ndarray_reshape[i])
    #print(index_sum_4_7)
    
    #後半は前半の何倍か
    several_times = int(index_sum_4_7) / int(index_sum_0_3)
    print("チェス盤の後半にある小麦の数は前半にある小麦の数の『{0}』倍です。".format(several_times))
    
    #print(board_ndarray_reshape[0])
    #print(board_ndarray_reshape[1])
    #print(board_ndarray_reshape[2])
    #print(board_ndarray_reshape[3])
    #print("-------------------")
    #print(np.sum(board_ndarray_reshape[0]))
    #print(np.sum(board_ndarray_reshape[1]))
    #print(np.sum(board_ndarray_reshape[2]))
    #print(np.sum(board_ndarray_reshape[3]))
    #print(np.sum(board_ndarray_reshape[4]))
    #print(np.sum(board_ndarray_reshape[5]))
    #print(np.sum(board_ndarray_reshape[6]))
    #print(np.sum(board_ndarray_reshape[7]))
    return

###----関数に縦8マス横8マスを代入----
choice_squaresnumber_komugi_chess(8, 8)
```

```python
%%timeit
#----インポート----
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np


#----引数の定義----
#n_Horizontal:横のマス数
#n_Vertical:縦のマス数

#---【ブロードキャスト_ver】n×mマスのチェス盤の小麦の数を算出する関数---
def choice_squaresnumber_komugi_chess_bc(n_Horizontal, n_Vertical):
    
    #マス数の計算（最初の１粒は置かないので、１を引かない）
    n_squares = int(n_Horizontal) * int(n_Vertical) 
    #print(n_squares)
    
    #64連続のndarray配列を作成
    indices_of_squares = np.arange(n_squares).astype(np.uint64)
    #print(indices_of_squares)
    
    #2の64乗を入れた
    board_list_bc = 2**indices_of_squares
    #print(board_list_bc)
    
    print("《ブロードキャストで計算した》{0}×{1}マスのチェス盤の小麦の数は 『{2}』です。\n".format(n_Horizontal, n_Vertical,board_list_bc[-1]))
    return


#----【ブロードキャスト_ver】関数に縦8マス横8マスを代入----
choice_squaresnumber_komugi_chess_bc(8, 8)
```
