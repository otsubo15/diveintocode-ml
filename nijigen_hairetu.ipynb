#### 【問題3】勾配を求める

```python
#----インポート----
import numpy as np

#X軸を0.1刻み格納したndarray。最後の値は表示されないので50.1としている
x_axis_ndarray = np.arange(-50,50.1,0.1)

#yの関数（ndarrayは直接割ったり足したりできる）
y_axis_ndarray = (((1/2) * x_axis_ndarray) + 1)

#ndarrayを結合するnp.stack()関数でaxis=1方向に結合
x_y_ndarray = np.stack([x_axis_ndarray, y_axis_ndarray], axis=1)

"""----軸の差分はの計算式ｰｰｰ
diff_a = a_ndarray[1:] - a_ndarray[:-1]
"""

#----Xの変化量----
diff_x = x_axis_ndarray[1:]-x_axis_ndarray[:-1]

#----Yの変化量----
diff_y = y_axis_ndarray[1:]-x_axis_ndarray[:-1]

#----勾配（slope）=yの変化量（diff_y）/xの変化量（diff_x）----
slope = diff_y / diff_x
print("勾配のndarrayは\n「{0}」\nです。".format(str(slope)))

```

#### 【問題5】Pythonの関数化

```python
#----インポート----
import matplotlib.pyplot as plt
import numpy as np


#①slope_and_graphs関数を実行（y = x ** 2, x_min=-50, x_max=50） について

###----関数①------------
def slope_and_graphs(x_min,x_max):
    
    ###----問題①の出力部------------
    
    #X軸を0.1刻み格納したndarray。
    x_axis_ndarray = np.arange(x_min,x_max,0.1)
    
    #yの関数（ndarrayは直接割ったり足したりできる）
    y_axis_ndarray = (x_axis_ndarray) ** 2 #★☆ここの関数を変える★☆
    
    ###----問題②の出力部------------
    
    #ndarrayを結合するnp.stack()関数でaxis=1方向に結合
    x_y_ndarray = np.stack([x_axis_ndarray, y_axis_ndarray], axis=1)
    
    ###----問題③の出力部------------
        
    #----Xの変化量----
    diff_x = x_axis_ndarray[1:]-x_axis_ndarray[:-1]
    
    #----Yの変化量----
    diff_y = y_axis_ndarray[1:]-x_axis_ndarray[:-1]
    
    #----勾配（slope）=yの変化量（diff_y）/xの変化量（diff_x）----
    slope = diff_y / diff_x
    
    return x_axis_ndarray, y_axis_ndarray, x_y_ndarray, slope
    ###----関数①ここまで------------



###----関数の呼び出し------------
x_axis_ndarray, y_axis_ndarray, x_y_ndarray, slope = slope_and_graphs(-50, 50)
print("１番目の関数を実行（y = x ** 2, x_min=-50, x_max=50） について")
print("\n")
print("《問題①の出力》")
print("①x_minからmax,yまで0.1間隔のxを表すndarray：\n{0} ".format(x_axis_ndarray)) 
print("\n")
print("②各要素に対するyを表すndarray：\n{0} ".format(y_axis_ndarray))
print("\n")
print("《問題②の出力》")
print(x_y_ndarray)
print("\n")
print("《問題③の出力》")
print("勾配のndarrayは\n「{0}」\nです。".format(str(slope)))

#----グラフ出力部----

plt.plot(x_axis_ndarray, y_axis_ndarray) #----y = x をグラフにplot
plt.show() #----グラフに出力


#----②slope_and_graphs関数を実行（y = 2 * x ** 2 + 2**x , x_min=-50, x_max=50） について

###----関数①------------
def slope_and_graphs(x_min,x_max):
    
    ###----問題①の出力部------------
    
    #X軸を0.1刻み格納したndarray。
    x_axis_ndarray = np.arange(x_min,x_max,0.1)
    
    #yの関数（ndarrayは直接割ったり足したりできる）
    y_axis_ndarray = 2 * x_axis_ndarray ** 2 + 2 ** x_axis_ndarray #★☆ここの関数を変える
    
    ###----問題②の出力部------------
    
    #ndarrayを結合するnp.stack()関数でaxis=1方向に結合
    x_y_ndarray = np.stack([x_axis_ndarray, y_axis_ndarray], axis=1)
    
    ###----問題③の出力部------------
        
    #----Xの変化量----
    diff_x = x_axis_ndarray[1:]-x_axis_ndarray[:-1]
    
    #----Yの変化量----
    diff_y = y_axis_ndarray[1:]-x_axis_ndarray[:-1]
    
    #----勾配（slope）=yの変化量（diff_y）/xの変化量（diff_x）----
    slope = diff_y / diff_x
    
    return x_axis_ndarray, y_axis_ndarray, x_y_ndarray, slope
    ###----関数①ここまで------------



###----関数の呼び出し------------
x_axis_ndarray, y_axis_ndarray, x_y_ndarray, slope = slope_and_graphs(-50, 50)
print("２番目の関数を実行（y = x ** 2, x_min=-50, x_max=50） について")
print("\n")
print("《問題①の出力》")
print("①x_minからmax,yまで0.1間隔のxを表すndarray：\n{0} ".format(x_axis_ndarray)) 
print("\n")
print("②各要素に対するyを表すndarray：\n{0} ".format(y_axis_ndarray))
print("\n")
print("《問題②の出力》")
print(x_y_ndarray)
print("\n")
print("《問題③の出力》")
print("勾配のndarrayは\n「{0}」\nです。".format(str(slope)))

#----グラフ出力部----

plt.plot(x_axis_ndarray, y_axis_ndarray) #----y = x をグラフにplot
plt.show() #----グラフに出力

#----③slope_and_graphs関数を実行（y = np.sin(x ** (1/2)) , x_min=0, x_max=50） について

###----関数①------------
def slope_and_graphs(x_min,x_max):
    
    ###----問題①の出力部------------
    
    #X軸を0.1刻み格納したndarray。
    x_axis_ndarray = np.arange(x_min,x_max,0.1)
    
    #yの関数（ndarrayは直接割ったり足したりできる）
    y_axis_ndarray = np.sin(x_axis_ndarray ** (1/2)) #★☆ここの関数を変える
    
    ###----問題②の出力部------------
    
    #ndarrayを結合するnp.stack()関数でaxis=1方向に結合
    x_y_ndarray = np.stack([x_axis_ndarray, y_axis_ndarray], axis=1)
    
    ###----問題③の出力部------------
        
    #----Xの変化量----
    diff_x = x_axis_ndarray[1:]-x_axis_ndarray[:-1]
    
    #----Yの変化量----
    diff_y = y_axis_ndarray[1:]-x_axis_ndarray[:-1]
    
    #----勾配（slope）=yの変化量（diff_y）/xの変化量（diff_x）----
    slope = diff_y / diff_x
    
    return x_axis_ndarray, y_axis_ndarray, x_y_ndarray, slope
    ###----関数①ここまで------------



###----関数の呼び出し------------
x_axis_ndarray, y_axis_ndarray, x_y_ndarray, slope = slope_and_graphs(0, 50)
print("３番目の関数を実行（y = np.sin(x ** (1/2)) , x_min=0, x_max=50） について")
print("\n")
print("《問題①の出力》")
print("①x_minからmax,yまで0.1間隔のxを表すndarray：\n{0} ".format(x_axis_ndarray)) 
print("\n")
print("②各要素に対するyを表すndarray：\n{0} ".format(y_axis_ndarray))
print("\n")
print("《問題②の出力》")
print(x_y_ndarray)
print("\n")
print("《問題③の出力》")
print("勾配のndarrayは\n「{0}」\nです。".format(str(slope)))

#----グラフ出力部----

plt.plot(x_axis_ndarray, y_axis_ndarray)#----y = x をグラフにplot
plt.show()#----グラフに出力
```
