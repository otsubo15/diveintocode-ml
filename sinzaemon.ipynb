【問題1】100日目の米粒の数

```python
import matplotlib.pyplot as plt
%matplotlib inline

#初日にもらう米を記載する
list_n_grains = []
list_total_grains = []
rice = int(1)
list_n_grains.append(rice)
list_total_grains.append(rice)
print("今日は{}日目です".format(str(1)))
print("「その日にもらう米の数」は{}粒です".format(rice))
print("「その日までに累計で貰う米の数」は{}粒です".format(sum(list_n_grains)))
print("------------------------------")

#for文でもらう米の数をリストに追加していく
for i in range(int(100-1)):
    rice = rice * 2
    list_n_grains.append(rice)
    list_total_grains.append(sum(list_n_grains))
    
    print("今日は{}日目です".format(str(i+2)))
    print("「その日にもらう米の数」は{}粒です".format(rice))
    print("「その日までに累計で貰う米の数」は{}粒です".format(sum(list_n_grains)))
    print("その日にもらう米の数リストは{}".format(list_n_grains))
    print("その日までに累計で貰う米の数リストは{}".format(list_total_grains))
    print("------------------------------")

#print(len(list_n_grains))
#print(len(list_total_grains))

# 「その日にもらう米の数」グラフを表示する。
plt.title("rice vol today gets")
plt.xlabel("days")
plt.ylabel("number of rice")
plt.plot(list_n_grains)
plt.show()

# 「その日までに累計で貰う米の数」グラフを表示する。
plt.title("total rice vol")
plt.xlabel("days")
plt.ylabel("number of rice")
plt.plot(list_total_grains, color='red')
plt.show()
```

【問題2】100日目以外の米粒の数

```python
def compute_sorori_shinzaemon(n_days=100): #日数を引数として入力しもらえる米数を数える関数
    
    #初日のもらえる米数について
    list_n_grains = []
    list_total_grains = []
    rice = int(1)
    list_n_grains.append(rice)
    list_total_grains.append(rice)
    print("今日は{}日目です".format(str(1)))
    print("「その日にもらう米の数」は{}粒です".format(rice))
    print("「その日までに累計で貰う米の数」は{}粒です".format(sum(list_n_grains)))
    print("------------------------------")
    
    #２日目以降のもらえる米数について（for文で繰り返し）    
    for i in range(int(n_days-1)):
        rice = rice * 2
        list_n_grains.append(rice)
        list_total_grains.append(sum(list_n_grains))

        print("今日は{}日目です".format(str(i+2)))
        print("「その日にもらう米の数」は{}粒です".format(rice))
        print("「その日までに累計で貰う米の数」は{}粒です".format(sum(list_n_grains)))
        print("その日にもらう米の数リストは{}".format(list_n_grains))
        print("その日までに累計で貰う米の数リストは{}".format(list_total_grains))
        print("------------------------------")

    # 「その日にもらう米の数」グラフを表示する。
    plt.title("rice vol today gets")
    plt.xlabel("days")
    plt.ylabel("number of rice")
    plt.plot(list_n_grains)
    plt.show()

    # 「その日までに累計で貰う米の数」グラフを表示する。
    plt.title("total rice vol")
    plt.xlabel("days")
    plt.ylabel("number of rice")
    plt.plot(list_total_grains, color='red')
    plt.show()
    return
###------------------------------

#コース受講日数は７月～１０月の合計１２３日なので
compute_sorori_shinzaemon(123)
```

【問題3】貰える米で何人が何日生活できるか

```python
"""
前提知識と必須入力項目について
関数名：Date_of_life（生きられる日数）
入力①米粒の数(grains)：問題２で算出したlist_total_grains[-1]を使います
入力②人数（Number)：デフォルトで14人にしておきます
出力：生活できる日数(Number_of_days_living)
一日に必要な米の量(Needed_rice)：6552（インターネットで調べたら一日茶碗２杯必要で、１杯3276粒でした。）
"""
def Date_of_life(grains = 1000000,Number = 14):
    Needed_rice = int(6552)
    Number_of_days_living = int(grains) / (Needed_rice * Number)
    print("貰える米{0}粒で{1}人が何日生活できるか,{2}日です。".format(int(grains),int(Number),int(Number_of_days_living)))
    return 
    
Date_of_life(grains = 10633823966279326983230456482242756607,Number = 14)


```
